Representaion  of   Police  catching    a   thief

```python
import numpy as np;
import matplotlib.pyplot as plt;
```

```python
t= np.linspace(0,40,10)
d_t=2.5*t+3
d_p=4*t+1
```

```python
fix,ax= plt.subplots()
plt.title("catching the thief")
plt.xlabel("time")
plt.ylabel("distance")
ax.plot(t,d_t,c='red')
ax.plot(t,d_p,c='green')
```
